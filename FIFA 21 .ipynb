### Importing data Fifa 21 dataset from Kaggle

```python
import pandas as pd
import numpy as np
```

```python
fifa_df = pd.read_csv("fifa21_dataset.csv")
```

```python
fifa_df.head()
```

```python
fifa_df.info()
```

As we check, we see that Height, weight and many other columns  are wrongly marked as object dataype

```python
df = fifa_df.convert_dtypes()
df.info()
```

```python
df["Hits"]
#As Hits has missing values, it was not converted to integer datatype
```

```python
df["Club"].head()
```

As we can see that the Club column contains the newline characters in it so we have to transform it 

```python
df["Club"].replace('\n',' ', regex = True, inplace = True)
```

```python
df["Club"].head()
```

```python
df.sort_values(["↓OVA"],ascending=True).head(20)
```

```python
fifa_df["Hits"].isnull().values.sum()
```

So, we now know that out of 18,979 players, 2595 players don't have a hit value.
Also, Loan date column will have null values because only a handful players are on loan at some club.

```python
loan_play=df[df['Loan Date End'].notnull()]
not_on_loan_play = df[df['Loan Date End'].isnull()]
```

List of 1013 players on loan with their name, club, age and potential

```python
loan_play[['Name', 'Club', 'Age', '↓OVA', 'POT']]
```

```python
not_on_loan_play[['Name', 'Club', 'Age', '↓OVA', 'POT']]
```

```python
x = loan_play["Age"].mean()
y = not_on_loan_play["Age"].mean()
print("The avg age of player on loan is ",x," whereas avg age of players not on loan is ", y )
```

## Therefore, This gives the insight that players out on loan are tend to be younger 

```python
total_players = df.groupby(['Nationality'])['Nationality'].count()
total_players.sort_values(ascending = False).head(5).plot.bar(title= "Top 5 countries with max no of players on Fifa 21",label = "Countries", ylabel= "No. of Players")
```

```python
no_hit = df[df["Hits"].isnull()]
no_hit.groupby(["Nationality"])["Nationality"].count().sort_values(ascending = False).head(5).plot.bar(title = "Top 5 countries with players having no hits")
```

```python
print("The avg age of player with no hits is ", no_hit["Age"].mean())
```

```python
print("The avg age of player with hits is ", df["Age"].mean())
```

On comparision of Top 5 countries with most players vs Top 5 countries with players with no hits, we show that spain, france, argentina moved out of top 5 positions when players with no hits were considered and countries, like, China PR, Saudi Arabia and Republic of Ireland replaced them. Meanwhile, England maintained it's top position and Germany slipped from 2nd to 4th.
On analysis of players with no hits, we found out that average age of players with no hits was way lower than players with hits available.
So, this raises a new question:

## Do China, Saudi Arabia and Ireland posses more younger talent than countries like Spain, France and Argentina? Do they have better future prospects?

```python
#We compare avg ages of players with no hit values among top nations
age= no_hit.groupby(['Nationality'])["Age"].mean()
age[["England", "China PR", "Saudi Arabia", "Germany",  "Spain", "France", "Argentina", "Republic of Ireland"]].sort_values()
```

```python
age[["England", "China PR", "Saudi Arabia", "Germany",  "Spain", "France", "Argentina"]].plot.line( )
```

```python
#We compare avg Potential of players with no hit values among top nations
pot = no_hit.groupby(['Nationality'])["POT"].mean()
pot[["England", "China PR", "Saudi Arabia", "Germany",  "Spain", "France", "Argentina", "Republic of Ireland"]].sort_values()
```

```python
pot[["England", "China PR", "Saudi Arabia", "Germany",  "Spain", "France", "Argentina"]].plot.line( )
```

We compared the average age of players with no hit values among top nations against the average age of players with no hit values. It gave us the conclusion that top football nations like Spain, England, Germany, France and Argentina have younger players in the no hit players data while nations like 
China PR, Ireland and Saudi have older players with low potential.
So, the answer for the above question would be No,they don't have more younger talents than other top nations.They rather have older players with low potential.
